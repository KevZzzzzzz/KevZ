```python
N=25# not recommend let N=100

braceletno = []
max_length = []

for mod in range(1,N+1):
    
    diffbracelet = set()# to make sure we not include every previous result
    
    for A in range(mod):
        for b in range(mod):
            i1 = A
            j1 = b
            k1 = (A+b) % mod
            
            m1 = mod**2
            m2 = 2*m1
            bracelet = np.empty(m2,dtype=int) # we using m2 here to ensure there are enough numbers to loop
            
            bracelet[0] = A
            bracelet[1] = b
            
            n = 2
            i = A 
            j = b
            
            while n<=m2-2:
                k=(i+j) % mod
                i,j = j,k
                bracelet[n] = k
                n = n+1
                
            m = 2
            x = 2
            while x <= m2-2:
                if bracelet[x] == i1 and bracelet[x+1] == j1 and bracelet[x+2] == k1:
                    break
                else:
                    m = m+1
                x = x+1
            
                
            diffbracelet.add(m)
          
    braceletno.append(len(diffbracelet))
    max_length.append(max(diffbracelet))  
    

#plotting
plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
plt.bar(range(1, N+1), braceletno)
plt.title('Number of Different Bracelets')
plt.xlabel('N')
plt.ylabel('Number of Different Bracelets')

plt.subplot(1, 3, 2)
plt.plot(range(1, N+1), max_length, marker='o')
plt.title('Max Possible Bracelet Length')
plt.xlabel('N')
plt.ylabel('Max Possible Bracelet Length')

plt.subplot(1, 3, 3)
plt.loglog(range(1, N+1), max_length, marker='o')
plt.title('Log-Log Plot of Max Possible Bracelet Length')
plt.xlabel('N')
plt.ylabel('Max Possible Bracelet Length')

plt.tight_layout()
plt.show()
```
