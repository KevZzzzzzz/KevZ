```python
import random

#hat game
def hatgame(n):
    number = list(range(1,n+1))
    
    while len(number)>1:
        i, j = random.sample(range(len(number)),2)
        diff = abs(number[i] - number[j])
        number.pop(max(i, j))  
        number.pop(min(i, j))
        number.append(diff)
    return number[0]

def repeat(k):
    result = []
    for l in range(k):  
        outcome = hatgame(n)  
        result.append(outcome)  
    return result 
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from scipy.special import comb

n = 2024
k = 5000

result = repeat(k)
count = Counter(result)

uniqueoutcome = list(count.keys())
occurrence = list(count.values())

plt.bar(uniqueoutcome,occurrence)
plt.xlabel('Result')
plt.ylabel('Occurrence')
plt.title(f'Occurrence of outcomes N={n}')
plt.show()
```
